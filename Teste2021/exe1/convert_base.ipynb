# Imports

```python
import rdflib 
from rdflib import Graph, Literal, RDF, URIRef, BNode
from rdflib.namespace import DC, DCTERMS, DOAP, FOAF, SKOS, OWL, RDF, RDFS, VOID, XMLNS, XSD
import os
```

# 1. Start Graph

```python
g = rdflib.Graph()
url = "http://di.uminho.pt/prc2021/teste"
g.bind("owl", OWL)
g.bind("rdfs", RDFS)
g.bind("rdf", RDF)
g.bind("xml", XMLNS)
g.bind("xsd", XSD)
```

# 2. Filling

## 2.1 Object Properties


```python
#exame_feito_em
g.add((
  rdflib.URIRef(url+"#exame_feito_em"),
  RDF.type,
  OWL.ObjectProperty
))

#pratica
g.add((
  rdflib.URIRef(url+"#pratica"),
  RDF.type,
  OWL.ObjectProperty
))
g.add((
  rdflib.URIRef(url+"#pratica"),
  OWL.inverseOf,
  rdflib.URIRef(url+"#é_praticada_por")
))

#inscrito_no_clube
g.add((
  rdflib.URIRef(url+"#inscrito_no_clube"),
  RDF.type,
  OWL.ObjectProperty
))
g.add((
  rdflib.URIRef(url+"#inscrito_no_clube"),
  RDFS.domain,
  rdflib.URIRef(url+"#Atleta")
))
g.add((
  rdflib.URIRef(url+"#inscrito_no_clube"),
  RDFS.range,
  rdflib.URIRef(url+"#Clube")
))

#é_praticada_por
g.add((
  rdflib.URIRef(url+"#é_praticada_por"),
  RDF.type,
  OWL.ObjectProperty
))
g.add((
  rdflib.URIRef(url+"#é_praticada_por"),
  RDFS.domain,
  rdflib.URIRef(url+"#Modalidade")
))
g.add((
  rdflib.URIRef(url+"#é_praticada_por"),
  RDFS.range,
  rdflib.URIRef(url+"#Atleta")
))

#modalidade_de
g.add((
  rdflib.URIRef(url+"#modalidade_de"),
  RDF.type,
  OWL.ObjectProperty
))
g.add((
  rdflib.URIRef(url+"#modalidade_de"),
  OWL.inverseOf,
  rdflib.URIRef(url+"#tem_a_modalidade")
))
g.add((
  rdflib.URIRef(url+"#modalidade_de"),
  RDFS.domain,
  rdflib.URIRef(url+"#Modalidade")
))
g.add((
  rdflib.URIRef(url+"#modalidade_de"),
  RDFS.range,
  rdflib.URIRef(url+"#Clube")
))

#tem_a_modalidade
g.add((
  rdflib.URIRef(url+"#tem_a_modalidade"),
  RDF.type,
  OWL.ObjectProperty
))

g.add((
  rdflib.URIRef(url+"#emd_com_modalidade"),
  RDF.type,
  OWL.ObjectProperty
))
```

## 2.2 Data Properties


```python
g.add((
  rdflib.URIRef(url+"#apelido"),
  RDF.type,
  OWL.DatatypeProperty
))
g.add((
  rdflib.URIRef(url+"#dataEMD"),
  RDF.type,
  OWL.DatatypeProperty
))
g.add((
  rdflib.URIRef(url+"#email"),
  RDF.type,
  OWL.DatatypeProperty
))
g.add((
  rdflib.URIRef(url+"#federado"),
  RDF.type,
  OWL.DatatypeProperty
))
g.add((
  rdflib.URIRef(url+"#género"),
  RDF.type,
  OWL.DatatypeProperty
))
g.add((
  rdflib.URIRef(url+"#idade"),
  RDF.type,
  OWL.DatatypeProperty
))
g.add((
  rdflib.URIRef(url+"#morada"),
  RDF.type,
  OWL.DatatypeProperty
))
g.add((
  rdflib.URIRef(url+"#nome_clube"),
  RDF.type,
  OWL.DatatypeProperty
))
g.add((
  rdflib.URIRef(url+"#nome_modalidade"),
  RDF.type,
  OWL.DatatypeProperty
))
g.add((
  rdflib.URIRef(url+"#nome_prórprio"),
  RDF.type,
  OWL.DatatypeProperty
))
g.add((
  rdflib.URIRef(url+"#resultado"),
  RDF.type,
  OWL.DatatypeProperty
))
```

## 2.3 Classes

```python
g.add((
  rdflib.URIRef(url+"#Atleta"),
  RDF.type,
  OWL.Class
))
g.add((
  rdflib.URIRef(url+"#Clube"),
  RDF.type,
  OWL.Class
))
g.add((
  rdflib.URIRef(url+"#Modalidade"),
  RDF.type,
  OWL.Class
))
g.add((
  rdflib.URIRef(url+"#EMD"),
  RDF.type,
  OWL.Class
))
```

## 2.4 Individuals

```python
import json
```

```python
with open('emd.json', encoding="utf-8") as json_file:
    data = json.load(json_file)
    for emd in data:
        #emd
        g.add((
        rdflib.URIRef(url+"#" + emd['_id']),
        RDF.type,
        OWL.NamedIndividual
        ))
        g.add((
        rdflib.URIRef(url+"#" + emd['_id']),
        RDF.type,
        rdflib.URIRef(url+"#EMD")
        ))
        g.add((
        rdflib.URIRef(url+"#" + emd['_id']),
        URIRef(url + "#dataEMD"),
        Literal(emd["dataEMD"])
        ))
        g.add((
        rdflib.URIRef(url+"#" + emd['_id']),
        URIRef(url + "#resultado"),
        Literal(emd["resultado"],datatype=XSD.boolean)
        ))
        g.add((
        rdflib.URIRef(url+"#" + emd['_id']),
        URIRef(url+"#emd_com_modalidade"),
        rdflib.URIRef(url+"#" + emd['modalidade'])
        ))

        #modalidade
        g.add((
        rdflib.URIRef(url+"#" + emd['modalidade']),
        RDF.type,
        OWL.NamedIndividual
        ))
        g.add((
        rdflib.URIRef(url+"#" + emd['modalidade']),
        RDF.type,
        rdflib.URIRef(url+"#Modalidade")
        ))
        g.add((
        rdflib.URIRef(url+"#" + emd['modalidade']),
        URIRef(url + "#nome_modalidade"),
        Literal(emd["modalidade"])
        ))


        #clube
        clube = emd['clube'].replace(" ", "_")
        g.add((
        rdflib.URIRef(url+"#" + clube),
        RDF.type,
        OWL.NamedIndividual
        ))
        g.add((
        rdflib.URIRef(url+"#" + clube),
        RDF.type,
        rdflib.URIRef(url+"#Clube")
        ))
        g.add((
        rdflib.URIRef(url+"#" + clube),
        URIRef(url + "#nome_clube"),
        Literal(emd["clube"])
        ))
        g.add((
        rdflib.URIRef(url+"#" + clube),
        URIRef(url+"#tem_a_modalidade"),
        rdflib.URIRef(url+"#" + emd['modalidade'])
        ))

        #atleta
        nome = emd["nome"]["primeiro"] + '_' + emd["nome"]["último"]
        g.add((
        rdflib.URIRef(url+"#" + nome),
        RDF.type,
        OWL.NamedIndividual
        ))
        g.add((
        rdflib.URIRef(url+"#" + nome),
        RDF.type,
        rdflib.URIRef(url+"#Atleta")
        ))
        g.add((
        rdflib.URIRef(url+"#" + nome),
        URIRef(url+"#exame_feito_em"),
        rdflib.URIRef(url+"#" + emd['_id'])
        ))
        g.add((
        rdflib.URIRef(url+"#" + nome),
        URIRef(url+"#inscrito_no_clube"),
        rdflib.URIRef(url+"#" + clube)
        ))
        g.add((
        rdflib.URIRef(url+"#" + nome),
        URIRef(url+"#pratica"),
        rdflib.URIRef(url+"#" + emd['modalidade'])
        ))
        g.add((
        rdflib.URIRef(url+"#" + nome),
        URIRef(url + "#apelido"),
        Literal(emd["nome"]["último"])
        ))
        g.add((
        rdflib.URIRef(url+"#" + nome),
        URIRef(url + "#email"),
        Literal(emd["email"])
        ))
        g.add((
        rdflib.URIRef(url+"#" + nome),
        URIRef(url + "#federado"),
        Literal(emd["federado"],datatype=XSD.boolean)
        ))
        g.add((
        rdflib.URIRef(url+"#" + nome),
        URIRef(url + "#género"),
        Literal(emd["género"])
        ))
        g.add((
        rdflib.URIRef(url+"#" + nome),
        URIRef(url + "#idade"),
        Literal(emd["idade"])
        ))
        g.add((
        rdflib.URIRef(url+"#" + nome),
        URIRef(url + "#morada"),
        Literal(emd["morada"])
        ))
        g.add((
        rdflib.URIRef(url+"#" + nome),
        URIRef(url + "#nome_próprio"),
        Literal(emd["nome"]["primeiro"])
        ))
```

# 3. Convert


```python
pr = g.serialize(format="turtle")
g.serialize(destination='output.ttl',format="turtle")
```
